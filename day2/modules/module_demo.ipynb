```python
x = np.linspace(0., 10., 200)
plt.figure()
colors = "rbcm"
line_styles = ["-", "--", "-."]
f = [1., 2., 5., 7, 8, 10]
for i in range(len(f)):
    c = colors[i % len(colors)]
    ls = line_styles[i % len(line_styles)]
    plt.plot(x, np.sin(2. * np.pi * f[i]/50 * x),
             color = c,
             linestyle = ls,
             label = "$f = {0}$".format(f[i]))
plt.grid()
plt.legend(loc =  "best")
plt.show()
```
