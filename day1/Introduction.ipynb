# Vers un Python "scientifique" ?

Des besoins spécifiques, des librairies adaptées

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib nbagg
```

```python
def amplitude(x , f = 1., a = 1.):
    return np.sin(2. * np.pi * f * x) * np.exp(-x / a)

x = np.linspace(0, 10., 300)
plt.figure()
plt.plot(x, amplitude(x, f = 2., a = 8.), label = "$f = 2$ $a = 5$")
plt.plot(x, amplitude(x, f = 1., a = 3.), label = "$f = 1$ $a = 3$")
plt.grid()
plt.legend()
plt.xlabel("Axe $x$")
plt.ylabel("Axe $y$")
plt.show()
```
